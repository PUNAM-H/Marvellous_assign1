# 1. What is the difference between local variables and global variables?

A local variable is declared inside a function and can only be used within that function.
It exists only while the function is executing.

A global variable is declared outside all functions and can be accessed from anywhere in the program.
It exists throughout the execution of the program.

```python
x = 100      # Global variable

def demo():
    y = 50   # Local variable
    print(y)

demo()
print(x)
```

# 2. Why should excessive use of global variables be avoided in large programs?


Global variables should be used carefully because:

They make the code difficult to understand.
Any function can change their value, causing unexpected errors.
They reduce code readability and maintainability.
They make debugging and testing more difficult.

Prefer local variables whenever possible.

# 4. What is the special variable __name__ in Python? Who assigns its value?

__name__ is a special built-in variable provided by Python.

# It tells us how a Python file is being executed:

If the file is run directly, __name__ is "__main__".
If the file is imported into another program, __name__ becomes the module's name.

The value of __name__ is assigned automatically by the Python interpreter.

It is commonly used to decide whether certain code should execute only when the file is run directly.

# 6. Can a function return another function? Explain conceptually.

Yes, Python allows a function to return another function because functions are treated as first-class objects.

# This means functions can be:

Assigned to variables
Passed as arguments
Returned from other functions

```python
def outer():
    def inner():
        return "Hello"
    return inner

f = outer()
print(f())
```

# 7. What is a module in Python? How is it different from a function?

A module in Python is a file with a .py extension that contains reusable Python code such as functions, classes, variables, and executable statements. Modules help organize code into separate files, making large applications easier to manage, maintain, and reuse.

Instead of writing the same code repeatedly, we can place it in a module and import it wherever needed. Python provides many built-in modules like math, random, datetime, and os, and we can also create our own custom modules.

A function, on the other hand, is a reusable block of code inside a module that performs a specific task. A module can contain one or many functions.

```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b


Now I can use this module in another program:
```

```python
import calculator

print(calculator.add(10, 5))


calculator.py is the module.
add() and subtract() are functions inside that module.
```

# 9. Why do we write the condition:

if __name__ == "__main__":

```python
is used to check whether a Python file is being executed directly or imported as a module into another Python program.

When Python executes a file directly, it automatically assigns the special variable __name__ the value "__main__".

If the same file is imported into another program, __name__ is assigned the module's name instead of "__main__".

Therefore, the code inside this if block runs only when the file is executed directly and does not run when the file is imported.
```

```python
def greet():
    print("Hello!")

if __name__ == "__main__":
    greet()
```

```python
Case 1: Execute directly
python demo.py

Output

Hello!
```

```python
Case 2: Import the file
import demo

Output

No output.

The function greet() becomes available, but it does not execute automatically.
```
