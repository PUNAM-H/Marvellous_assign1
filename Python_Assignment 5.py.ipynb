# 1. What are bytes in Python? Why are they immutable?

```python
Bytes are a sequence of binary data (0–255). They are mainly used to store binary files like images, audio, and videos.

Why immutable?
Once a bytes object is created, its values cannot be changed.
```

```python
b = bytes([65, 66, 67])
print(b)
```

# 8. What is a set in Python?
A Set in Python is a built-in collection data type used to store multiple unique values. It is unordered, meaning the elements do not have a fixed order, and it does not allow duplicate values. A set is mutable, so we can add or remove elements after it is created. Sets are mainly used when we need to store unique data and perform fast operations like searching, union, intersection, and removing duplicates.

```python
Stores unique elements only.
Does not allow duplicates.
Unordered collection.
Mutable (elements can be added or removed).
Written using curly braces {}.
```

# 10. What is Dynamic Typing in Python?
Python is dynamically typed because we do not need to declare the data type of a variable.
Python automatically decides the data type at runtime.

```python
a = 10
print(type(a))

a = 10.5
print(type(a))

a = "Python"
print(type(a))
```
