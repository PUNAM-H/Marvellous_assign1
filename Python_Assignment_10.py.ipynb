# 2) Write a program which accepts one number and prints the sum of first N natural numbers.

Example:

Input : 5
Output: 15

```python
num = int(input("Enter a number: "))

sum = 0

for i in range(1, num + 1):
    sum += i
    
print("Sum =", sum)
```

```python
num = int(input("Enter a number: "))

sum = 0
i = 1

while i <= num:
    sum += i
    i += 1

print("Sum =", sum)
```

```python
def Sum(no):
    total = 0
    for i in range(1, no + 1):
        total += i
    return total

def main():
    num = int(input("Enter a number: "))
    print("Sum =", Sum(num))

if __name__ == "__main__":
    main()
```

```python
def Sum(no):
    total = 0
    i = 1
    while i <= no:
        total += i
        i += 1
    return total

def main():
    num = int(input("Enter a number: "))
    print("Sum =", Sum(num))

if __name__ == "__main__":
    main()
```

# 3) Write a program which accepts one number and prints the factorial of that number.

Example:

Input : 5
Output: 120

```python
num = int(input("Enter a number: "))

fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial =", fact)
```

```python
num = int(input("Enter a number: "))

fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial =", fact)
```

```python
def Factorial(no):
    fact = 1
    for i in range(1, no + 1):
        fact *= i
    return fact

def main():
    num = int(input("Enter a number: "))
    print("Factorial =", Factorial(num))

if __name__ == "__main__":
    main()
```

```python
def Factorial(no):
    fact = 1
    i = 1
    while i <= no:
        fact *= i
        i += 1
    return fact

def main():
    num = int(input("Enter a number: "))
    print("Factorial =", Factorial(num))

if __name__ == "__main__":
    main()
```

# 4) Write a program which accepts one number and prints all even numbers till that number.

Example:

Input : 10
Output: 2 4 6 8 10

```python
num = int(input("Enter a number: "))

for i in range(2, num + 1, 2):
    print(i, end=" ")
```

```python
num = int(input("Enter a number: "))

i = 2

while i <= num:
    print(i, end=" ")
    i += 2
```

```python
def Even(no):
    for i in range(2, no + 1, 2):
        print(i, end=" ")

def main():
    num = int(input("Enter a number: "))
    Even(num)

if __name__ == "__main__":
    main()
```
