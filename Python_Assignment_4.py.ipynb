# Q2. What is a Set in Python?

A Set is a built-in data type in Python used to store a collection of unique and unordered elements. It automatically removes duplicate values and allows efficient operations like union, intersection, and difference.
Features of Set:
Stores only unique values.
Does not allow duplicate elements.
Unordered collection.
Mutable (elements can be added or removed).
Defined using curly braces {} or the set() function.

```python
numbers = {10, 20, 30, 20, 40}

print(numbers)
```

# Q3. Write a Python program to create a set and perform add and remove operations.

```python
fruits = {"Apple", "Banana", "Mango"}

print("Original Set:", fruits)

fruits.add("Orange")
print("After Adding:", fruits)

fruits.remove("Banana")
print("After Removing:", fruits)
```

# Q5. Write a Python program to find the union and intersection of two sets.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union:", A.union(B))
print("Intersection:", A.intersection(B))
```

# Q8. What is the range data type in Python? How is it different from a list of numbers?
The range data type is a built-in Python sequence used to generate a sequence of numbers. It is commonly used in for loops. A range object generates numbers only when needed, making it memory efficient.

# Range:                                                          List:
Generates numbers on demand.                            Stores all numbers in memory.
Uses less memory.                                       Uses more memory.
Mainly used for looping.                                Used to store and modify data.
Immutable.	                                            Mutable.

# Q10. What is the difference between range(1, 10) and range(1, 10, 2)? Explain the parameters.

start: The starting value (included).
stop: The ending value (not included).
step: The increment or difference between consecutive numbers.

```python
1. range(1, 10)
Starts from 1
Ends before 10
Default step is 1
```

```python
print(list(range(1, 10)))
```

```python
2. range(1, 10, 2)
Starts from 1
Ends before 10
Increases by 2
```
