# Q1. Write a program which accepts one number and checks whether it is prime or not.

```python
num=int(input("Enter the number: "))
for i in range (2,num):
    if num % i == 0:
        print("This number is Not Prime")
        break
        
else:
    print("This number is Prime ")
```

```python
def ChkPrime(num):
    for i in range (2,num):
        
        if num % i==0:
            print("Not Prime")
            
            break
            
    else:
        print("Prime")
        
            
def main():
    no=int(input("Enter the the number:"))
    ChkPrime(no)
    
if __name__ == "__main__":
    main()
```

# Q2. Write a program which accepts one number and prints count of digits in that number.

```python
def CountDigits(no):
    count = 0

    while no > 0:
        count = count + 1
        no = no // 10

    return count


def main():
    num = int(input("Enter the number: "))
    print(CountDigits(num))


if __name__ == "__main__":
    main()
```

# Q3. Write a program which accepts one number and prints sum of digits.

```python
def SumDigits(no):
    sum = 0

    while no > 0:
        sum = sum + (no % 10)
        no = no // 10

    return sum


def main():
    num = int(input("Enter the number: "))
    print(SumDigits(num))


if __name__ == "__main__":
    main()
```

# Q4. Write a program which accepts one number and prints reverse of that number.

```python
def Reverse(no):
    rev = 0

    while no > 0:
        rev = rev * 10 + (no % 10)
        no = no // 10

    return rev


def main():
    num = int(input("Enter the number: "))
    print(Reverse(num))


if __name__ == "__main__":
    main()
```

# Q5. Write a program which accepts one number and checks whether it is palindrome or not.

```python
def Palindrome(num):
    if str(num) == str(num)[::-1]:
        print("Palindrome")
    else:
        print("Not Palindrome")


def main():
    num = int(input("Enter the number: "))
    Palindrome(num)


if __name__ == "__main__":
    main()
```
