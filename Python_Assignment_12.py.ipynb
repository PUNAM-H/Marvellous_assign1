# 1) Write a program which accepts one character and checks whether it is vowel or consonant.

```python
character=input("Enter the character: ")
if character in "aeiouAEIOU":
    print("vowel")
else:
    print("consonant")
```

# 2)  Write a program which accepts one number and prints its factors.

```python
num= int(input("Enter the number: "))
for i in range(1,num+1):
    if num % i == 0:
        print(i)
```

```python
def Factors(num):
    for i in range(1,num+1):
        if num % i == 0:
            print(i)
    
def main():
    number= int(input("Enter the number: "))
    Factors(number)
    
if __name__ == "__main__":
    main()
```

# 3 )Write a program which accepts two numbers and prints addition, subtraction, multiplication and division.

```python
def Arithmetic(no1,no2):
    print("Addition:",no1 + no2)
    print("SUbtraction:",no1 - no2)
    print("Multiplication:",no1 * no2)
    print("Division:",no1 / no2)
    
def main():
    num1=int(input("Enter the number first:"))
    num2=int(input("Enter the number Second:"))
    Arithmetic(num1,num2)
    
if __name__ == "__main__":
    main()
```

# 4) Write a program which accepts one number and prints that many numbers starting from 1.

```python
def Display(no):
    for i in range(1,no+1):
        print(i,end = " ")
    
    
def main():
    num=int(input("Enter the number first:"))
    Display(num)

if __name__ == "__main__":
    main()
```

# 5) Write a program which accepts one number and prints that many numbers in reverse order.

```python
def Reverse(no):
    for i in range(no,0,-1):
        print(i,end = " ")

def main():
    
    num=int(input("Enter the number first:"))
    Reverse(num)

           
if __name__ == "__main__":
    main()
```
