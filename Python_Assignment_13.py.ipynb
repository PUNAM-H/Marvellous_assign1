# Q1. Write a program to accept length and width of rectangle and print area.

```python
def Area(length,Width):
    return length * Width

def main():
    l=int(input("Enter the length: "))
    w=int(input("Enter the WIdth: "))
    print("Area of rectangle: ",Area(l,w))
    
    
if __name__ == "__main__":
    main()
```

# Q2. Write a program to accept radius of circle and print area.

```python
def AreaC(Radius):
    return 3.14 * Radius * Radius

def main():
    Radius=int(input("Enter the Radius: "))
    print("Area of Circle: ",AreaC(Radius))
    
    
if __name__ == "__main__":
    main()
```

# Q3. Write a program to check whether a number is perfect or not.

```python
def Perfect(no):
    sum = 0

    for i in range(1, no):
        if no % i == 0:
            sum = sum + i

    if sum == no:
        print("Perfect Number")
    else:
        print("Not Perfect Number")

def main():
    num = int(input("Enter number: "))
    Perfect(num)

if __name__ == "__main__":
    main()
```

# Q4. Write a program to print binary equivalent of a number.

```python
def Binary(no):
    print(bin(no))

def main():
    num = int(input("Enter number: "))
    Binary(num)

if __name__ == "__main__":
    main()
```

# Q5. Write a program to accept marks and display grade.

```python
def Grade(marks):
    if marks >= 75:
        print("Distinction")
    elif marks >= 60:
        print("First Class")
    elif marks >= 50:
        print("Second Class")
    else:
        print("Fail")

def main():
    m = int(input("Enter marks: "))
    Grade(m)

if __name__ == "__main__":
    main()
```
