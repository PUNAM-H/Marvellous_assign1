# 2. Difference between a = 10, b = 10 and a = [10], b = [10]



```python
# For integers, memory address may be same.
a=10
b=10
print(id(a))
print(id(b))

#For lists, memory address is different because both are separate objects.
a=[10]
b=[10]
print(id(a))
print(id(b))
```

# 3. What does id() function return?
id() returns the memory address (unique identity) of an object. Same value may or may not have same id().


```python
num=10
print(id(num))
```

# 4. Purpose of getsizeof()
getsizeof() shows memory size of object. Different data types use different memory.

```python
import sys
a = 10
b = "Hello"

print(sys.getsizeof(a))
print(sys.getsizeof(b))
```
